```python
import pandas as pd
import numpy as numpy
import sklearn.preprocessing as preprocessing
import sklearn.model_selection
```

```python
import pickle 

def save_pickle(obj, filename):
    with open("data/" + filename, 'wb') as f:
        pickle.dump(obj, f)

def load_pickle(filename):
    with open("data/" + filename, 'rb') as f:
        return pickle.load(f)
```

```python
data = pd.read_csv('dataset.csv')
data.head()
```

```python
drop_col = ['Symptom_12', 'Symptom_13', 'Symptom_14', 'Symptom_15', 'Symptom_16', 'Symptom_17']
data = data.drop(drop_col, axis=1)
data
```

```python
save_pickle(drop_col,'drop_col.pkl')  
```

```python
data['Disease'].unique()
```

```python
columns = data.columns[1:]
columns
```

```python
nul_replace = 'No additional Symtoms'
data.fillna(nul_replace, inplace=True)
save_pickle(nul_replace, 'nul_replace.pkl')
```

```python
x = ' sdfs  sdf  sdf s'
x= x.strip()
x
```

```python
for col in data.columns:
    data[col] = data[col].str.strip()
```

```python

symptoms = pd.unique(data[columns].values.ravel())

symptoms
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
symp = label_encoder.fit_transform(symptoms)
symp.sort()
symp


```

```python
save_pickle(label_encoder,'label_encoder.pkl')  
```

```python
encodedSymp = dict(zip(symptoms, symp))
encodedSymp
save_pickle(encodedSymp,'encodedSymp.pkl')  
```

```python
for col in columns:
    data[col] = data[col].map(encodedSymp)
```

```python
len(data['Disease'].unique())
```

```python
data
```

```python
x= data.drop('Disease', axis=1)
y= data['Disease']
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
save_pickle(sc,'StandardScaler.pkl')  
```

```python
x.iloc[505]
```

```python
x_scaled[505]
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42,max_depth=20)
model.fit(x_train, y_train)
```

```python
save_pickle(model,'model.pkl')  
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, precision_score
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:", confusion_matrix(y_test, y_pred))
print("Precision score:", precision_score(y_test, y_pred,average='weighted'))
model_score = model.score(x_test, y_pred)

print("Model score:", model_score)
```
